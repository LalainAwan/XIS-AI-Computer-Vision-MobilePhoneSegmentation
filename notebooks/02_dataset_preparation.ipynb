```python
import zipfile
import os

zip_path = "/content/drive/MyDrive/XIS_AI_Project/MobilePhoneSegmentation.v1i.coco-segmentation.zip"

print("File exists:", os.path.exists(zip_path))

extract_dir = "/content/mobile_phone_dataset"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_dir)

print("Dataset extracted successfully!")
```

```python
import os

train_json = "/content/mobile_phone_dataset/train/_annotations.coco.json"
valid_json = "/content/mobile_phone_dataset/valid/_annotations.coco.json"
test_json  = "/content/mobile_phone_dataset/test/_annotations.coco.json"

print("Train JSON exists:", os.path.exists(train_json))
print("Valid JSON exists:", os.path.exists(valid_json))
print("Test JSON exists:", os.path.exists(test_json))
```

```python
import json

with open(train_json, "r") as f:
    train_data = json.load(f)

with open(valid_json, "r") as f:
    valid_data = json.load(f)

with open(test_json, "r") as f:
    test_data = json.load(f)

print("COCO annotations loaded successfully!")
```

```python
#dataset statistics
print("TRAIN IMAGES :", len(train_data["images"]))
print("VALID IMAGES :", len(valid_data["images"]))
print("TEST IMAGES  :", len(test_data["images"]))

print("\nCATEGORIES:")

for cat in train_data["categories"]:
    print(cat)
```

```python
category_counts = {}

for ann in train_data["annotations"]:
    cat_id = ann["category_id"]
    category_counts[cat_id] = category_counts.get(cat_id, 0) + 1

print(category_counts)
```
